```python
# FPCA 降維
import FDApy
from FDApy.preprocessing.dim_reduction.fpca import UFPCA
from FDApy.representation.functional_data import DenseFunctionalData, DenseArgvals, DenseValues
```

```python
data = combined.T.astype(np.float64)
argvals = {'input_dim_0' : np.array(range(1, 2208+1))}
values = np.array(data)
fd = DenseFunctionalData(argvals, values)
ufpca = UFPCA(n_components=0.85, method='covariance')
ufpca.fit(fd)
scores_fpca = ufpca.transform(fd, method='NumInt')
```

```python
# 降維後的資料(score)
pd.DataFrame(scores_fpca)
```

```python
# 接著用 t-sne 降維
from sklearn.manifold import TSNE
import random
```

```python
# 最後降至 2 維
X_embedded = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=3, random_state=500).fit_transform(scores_fpca)
scores_fpca_tsne = pd.DataFrame(X_embedded)
scores_fpca_tsne
```

```python
# 接著用 K-means 分群

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# 發現當分 3 群時，分群效果最佳

silhouette_avg = []
for i in range(2,11):
    kmeans_fit = KMeans(n_clusters = i, random_state=0, n_init="auto").fit(scores_fpca_tsne)
    silhouette_avg.append(silhouette_score(scores_fpca_tsne, kmeans_fit.labels_))
plt.plot(range(2,11), silhouette_avg)
```

```python
kmeans = KMeans(n_clusters=3, random_state=0, n_init="auto").fit(scores_fpca_tsne)
kmeans_fit = KMeans(n_clusters = 3, random_state=0, n_init="auto").fit(scores_fpca_tsne)
print(silhouette_score(scores_fpca_tsne, kmeans_fit.labels_))
kmeans.labels_
```

```python
# 將降維後的資料分三群畫出來

random_state = 0

common_params = {
    "n_init": "auto",
    "random_state": random_state,
}

y_pred = KMeans(n_clusters=3, **common_params).fit_predict(scores_fpca_tsne)
plt.scatter(scores_fpca_tsne.iloc[:, 0], scores_fpca_tsne.iloc[:, 1], c=y_pred)
plt.title("Optimal Number of Clusters")
plt.show()
```

```python
# 計算兩座標距離
import pandas as pd
import numpy as np

def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate the great-circle distance between two points
    on the Earth's surface given their latitude and longitude
    in decimal degrees.
    """
    R = 6371  # Earth radius in kilometers

    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    # Differences in coordinates
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    # Haversine formula
    a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    c = 2 * np.arcsin(np.sqrt(a))

    # Distance in kilometers
    distance = R * c

    return distance
```
